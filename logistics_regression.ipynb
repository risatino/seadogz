```python
df = pd.merge(catapult_data, injury_data, 
              how='outer', left_on=['id', 'date'], 
              right_on=['id', 'injury_date'], left_index=False, 
              right_index=False, sort=False, suffixes=('_x', '_y'), 
              copy=True, indicator=False)

df["injury"].sum()
```

```python
df = df.fillna(value = 0)
#pd.set_option('display.max_rows', None)
df
```

```python
df = df.drop(['id', 'rhie_effort_duration_max.1', 'date','average_duration','rhie_effort_duration_max', 'rhie_effort_duration_min', 'injury_date'], axis=1)
```

```python
binary_df = pd.get_dummies(df)
binary_df.head()
binary_df.to_csv("binary_logistics_data.csv")
```

```python
X = binary_df.drop("injury", axis=1)
y = binary_df["injury"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=25, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
